### Jenis-Jenis Data
1. Data Kuantitatif: data yang menyatakan besaran, jumlah, atau jangkauan tertentu. Biasanya, terdapat unit pengukuran yang terkait dengan data, misalnya meter dalam pengukuran tinggi badan seseorang.
2. Data Kualitatif: data yang mendekati dan mencirikan, serta dapat diamati dan dicatat. Tipe data ini bersifat non-numerik dan dikumpulkan melalui metode observasi, wawancara, focus group, dan metode serupa
3. Data Interval: tipe data yang diukur sepanjang skala, di mana setiap titiknya ditempatkan pada jarak yang sama satu sama lain. Data interval selalu muncul dalam bentuk angka atau nilai numerik dengan jarak antara dua titik distandarisasi dan sama. Contoh dari data ini adalah pengukuran
4. Data Rasio: jenis data numerik yang bersifat kuantitatif. Data rasio yang dikumpulkan pada skala rasio memiliki jarak yang sama antara nilai-nilai yang berdekatan. Contoh dari data ini adalah skala
5. Data Ordinal: diklasifikasikan ke dalam variabel yang memiliki urutan peringkat alami, tetapi jarak antar kategori tidak diketahui. Misalnya, variabel “frekuensi latihan fisik” dapat dibagi dalam beberapa kategori, seperti tidak pernah, jarang, kadang-kadang, sering, dan selalu.

## Menangani outlier dengan variable

Outlier adalah observasi yang jauh dari nilai-nilai lain dalam dataset. Mereka bisa menjadi hasil dari kesalahan pengukuran, variabilitas alami dalam populasi, atau bahkan dapat menunjukkan informasi penting atau anomali dalam data

Dampak outlier terhadap analisis data dapat signifikan, termasuk:

1. **Pengaruh terhadap Statistik Deskriptif**: Outlier dapat menyebabkan pergeseran yang signifikan dalam rata-rata, median, dan standar deviasi, yang dapat mengaburkan pemahaman tentang karakteristik sebenarnya dari dataset.
2. **Pengaruh terhadap Model Statistik**: Outlier dapat mempengaruhi performa model statistik, seperti regresi linear, dengan menyebabkan model untuk salah memahami hubungan antara variabel. Ini dapat mengakibatkan model yang tidak akurat dan prediksi yang buruk.
3. **Pengaruh terhadap Kesimpulan dan Keputusan Bisnis**: Kesimpulan atau keputusan yang dibuat berdasarkan data yang mengandung outlier dapat menjadi tidak akurat, mengarah pada keputusan yang buruk atau tidak efektif dalam konteks bisnis.

Outlier dapat dilihat menggunakan visualisasi Box plot. Box plot adalah metode visualisasi yang digunakan untuk meringkas distribusi data numerik dengan menggambarkan kuartil data serta menampilkan outlier potensial.

## Menangani Missing Values

Missing values, atau nilai yang hilang, dapat memiliki dampak signifikan pada akurasi analisis data, terutama saat menggunakan algoritma pembelajaran mesin. Pengaruhnya dapat bervariasi tergantung pada seberapa banyak data yang hilang dan pada cara penanganan yang diterapkan. Berikut adalah beberapa pengaruh utama dari missing values pada akurasi:

1. **Pengurangan Ukuran Sampel**: Missing values dapat mengurangi ukuran sampel yang tersedia untuk analisis, yang dapat mengurangi akurasi model. Semakin banyak nilai yang hilang, semakin banyak informasi yang hilang, yang dapat memengaruhi kemampuan model untuk mengenali pola dan membuat prediksi yang akurat.
2. **Bias dalam Analisis**: Jika missing values tidak ditangani dengan benar, ini dapat menyebabkan bias dalam analisis. Jika pola atau kecenderungan tertentu terkait dengan data yang hilang, ini dapat memengaruhi hasil analisis secara keseluruhan.
3. **Ketidakakuratan dalam Estimasi Statistik**: Jika missing values tidak ditangani, ini dapat mengakibatkan perkiraan statistik yang tidak akurat, seperti rata-rata, standar deviasi, atau korelasi antar variabel. Hal ini dapat memengaruhi interpretasi analisis dan membuat kesimpulan yang salah.
4. **Distorsi pada Pola Data**: Missing values dapat menyebabkan distorsi pada pola data yang sebenarnya, terutama jika pola tersebut berkaitan dengan data yang hilang. Ini dapat mengakibatkan model yang dilatih pada data yang tidak mewakili populasi secara keseluruhan.


```python
dataframe.isnull().sum()
```

Feature Income memiliki data yang missing sejumlah 24 buah data. Untuk mengisi kekosongan data tersebut, data tersebut akan diisi oleh mean (rata-rata) dari Feature Income menggunakan metode *fillna*

```python
dataframe["Income"] = dataframe["Income"].fillna(dataframe["Income"].mean())
```

```python
dataframe.isnull().sum()
```

## Menangani data categorial

### Pengertian data categorial

Data categorial adalah jenis data yang mewakili kategori atau label, bukan nilai numerik yang kontinu. Data ini terdiri dari sejumlah kategori atau kelompok yang berbeda, yang sering kali mewakili atribut atau karakteristik tertentu dari suatu objek. Data kategorial tidak dapat diurutkan secara berurutan, tetapi hanya dapat dikategorikan ke dalam grup yang berbeda. Contoh umum dari data kategorial termasuk jenis kelamin, status pernikahan, kelas sosial, atau jenis produk.


### Pengaruh data categorial terhadap performa akurasi

1. **Pengkodean Data Kategorial**:
Data kategorial perlu dikodekan menjadi format yang dapat dipahami oleh model. Teknik pengkodean seperti One-Hot Encoding atau Label Encoding sering digunakan untuk mengubah data kategorial menjadi format yang dapat diolah oleh algoritma pembelajaran mesin.
2. **Penanganan Data Hilang**:
Kehadiran nilai yang hilang dalam data kategorial dapat memengaruhi akurasi model. Oleh karena itu, perlu adanya strategi penanganan nilai hilang yang tepat, seperti menghapus baris yang mengandung nilai null atau mengisi nilai tersebut dengan strategi tertentu, agar tidak mengganggu kinerja model.
3. **Seleksi Fitur**:
Sebelum memasukkan data kategorial ke dalam model, penting untuk memilih fitur yang paling relevan dan berpengaruh dalam memprediksi target. Seleksi fitur yang buruk dapat mengurangi akurasi model.
4. **Pemilihan Model yang Tepat**:
Beberapa model pembelajaran mesin mungkin lebih cocok untuk menangani data kategorial daripada yang lain. Oleh karena itu, pemilihan model yang tepat berdasarkan jenis data dan karakteristik masalah adalah penting untuk meningkatkan akurasi prediksi.

## Normalisasi data

Dalam konteks data ilmiah dan analisis data, normalisasi mengacu pada proses mengubah nilai-nilai dari berbagai fitur dalam dataset ke skala umum, sehingga setiap fitur memiliki rentang nilai yang serupa. Normalisasi penting karena beberapa algoritma pembelajaran mesin memerlukan data yang telah dinormalisasi untuk bekerja secara efisien dan memberikan hasil yang akurat.

Tujuan normalisasi adalah untuk memastikan bahwa setiap fitur memiliki dampak yang seimbang pada analisis dan model yang sedang dibangun. Hal ini membantu mencegah fitur dengan rentang nilai yang besar mendominasi model, sementara fitur dengan rentang nilai yang lebih kecil diabaikan. Dengan normalisasi, algoritma pembelajaran mesin dapat belajar dari setiap fitur dengan cara yang seimbang, sehingga menghasilkan model yang lebih akurat dan andal.

Jenis-Jenis normalisasi:
1. **Min-Max Scaling**: Ideal digunakan ketika Anda memiliki pemahaman yang jelas tentang rentang nilai maksimum dan minimum dari fitur Anda, dan Anda ingin menormalkan data ke rentang spesifik, misalnya 0 hingga 1. Metode ini cocok untuk data yang tidak memiliki distribusi normal atau ketika Anda ingin mempertahankan interpretasi asli dari data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_minmax = MinMaxScaler()

# Contoh DataFrame
data = {
    "Feature1": [100, 200, 300, 400, 500],
    "Feature2": [10, 20, 30, 40, 50],
    "Feature3": [5, 10, 15, 20, 25],
}

df = pd.DataFrame(data)

# Min-Max Scaling pada DataFrame
df_scaled = scaler_minmax.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)

# Menampilkan DataFrame setelah scaling
print("DataFrame setelah Min-Max Scaling:")
print(df_scaled)
```

2. **Z-Score Normalization**: Cocok digunakan ketika Anda ingin memastikan bahwa data Anda memiliki rata-rata nol dan simpangan baku satu. Metode ini berguna ketika data Anda memiliki distribusi normal atau mendekati distribusi normal, dan Anda ingin menghilangkan perbedaan skala yang besar antara fitur-fitur yang berbeda. Z-Score Normalization juga cocok untuk penggunaan algoritma yang bergantung pada asumsi normalitas data.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Contoh DataFrame
data = {
    "Feature1": [100, 200, 300, 400, 500],
    "Feature2": [10, 20, 30, 40, 50],
    "Feature3": [5, 10, 15, 20, 25],
}


df = pd.DataFrame(data)

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Z-Score Normalization pada DataFrame
df_normalized = scaler.fit_transform(df)
df_normalized = pd.DataFrame(df_normalized, columns=df.columns)

# Menampilkan DataFrame setelah normalisasi
print("DataFrame setelah Z-Score Normalization:")
print(df_normalized)
```

3. **Normalization by Decimal Scaling**: Digunakan ketika Anda ingin melakukan normalisasi dengan menggeser koma desimal dari setiap nilai fitur, sehingga nilai-nilai tersebut jatuh dalam rentang tertentu, seperti antara -1 dan 1. Metode ini berguna ketika Anda ingin menjaga signifikansi angka dan mengurangi kompleksitas perhitungan. Normalization by Decimal Scaling cocok untuk kasus di mana interpretasi data sangat penting dan Anda ingin mempertahankan representasi numerik asli.

```python
# Contoh DataFrame
data = {
    "Feature1": [100, 200, 300, 400, 500],
    "Feature2": [10, 20, 30, 40, 50],
    "Feature3": [5, 10, 15, 20, 25],
}

df = pd.DataFrame(data)

# Mencari jumlah digit maksimal dalam kolom
max_digits = df.applymap(lambda x: len(str(x))).max()

# Menghitung skala
scale = 10 ** (max_digits - 1)

# Normalisasi by Decimal Scaling pada DataFrame
df_normalized = df / scale

# Menampilkan DataFrame setelah normalisasi
print("DataFrame setelah Normalization by Decimal Scaling:")
print(df_normalized)
```

5. Histogram: Bentuk histogram distribusi normal menyerupai lonceng, dengan puncaknya di tengah dan ekornya yang menyebar ke kiri dan kanan.

## Reduksi Fitur

Reduksi fitur dalam konteks data ilmiah atau data science mengacu pada proses mengurangi jumlah fitur atau variabel yang digunakan dalam analisis data. Tujuan utama dari reduksi fitur adalah untuk menghilangkan fitur yang tidak relevan atau redundan, sehingga meningkatkan efisiensi komputasi, mengurangi overfitting, dan meningkatkan interpretabilitas model.

Pengaruh reduksi fitur pada performa tergantung pada cara reduksi fitur dilakukan dan karakteristik data yang sedang ditangani. Berikut adalah beberapa pengaruh umum dari reduksi fitur pada performa:

1. **Meningkatkan Efisiensi Komputasi**: Dengan mengurangi jumlah fitur, waktu komputasi yang dibutuhkan untuk melatih model dapat dikurangi, sehingga mempercepat proses analisis data.
2. **Mengurangi Overfitting**: Reduksi fitur dapat membantu mengurangi kompleksitas model, yang dapat mengurangi risiko overfitting, yaitu ketika model terlalu rumit sehingga cocok dengan noise atau variabilitas acak dalam data pelatihan.
3. **Meningkatkan Interpretabilitas Model**: Dengan menghilangkan fitur yang tidak relevan, model yang dihasilkan menjadi lebih mudah dipahami dan diinterpretasikan. Hal ini dapat membantu pemahaman yang lebih baik tentang hubungan antara fitur yang tersisa dan target yang ingin diprediksi.
4. **Mengurangi Dimensi Ruang Fitur**: Reduksi fitur juga dapat membantu mengurangi dimensi ruang fitur, terutama ketika Anda bekerja dengan dataset berdimensi tinggi. Hal ini dapat membantu dalam visualisasi data dan mencegah "kutukan dimensi" di mana kinerja model menurun seiring peningkatan dimensi data.

### Principal Component Analysis (PCA) untuk reduksi fitur
Principal Component Analysis (PCA) adalah salah satu metode yang umum digunakan untuk reduksi fitur dalam analisis data. PCA bekerja dengan mentransformasi dataset asli ke dalam sistem koordinat baru di mana sumbu utamanya adalah komponen utama (principal components) dari dataset tersebut. Komponen utama ini dipilih sedemikian rupa sehingga mereka mengandung sebagian besar variasi dalam data.

Dengan menggunakan PCA, dimensi data dapat dikurangi dengan memilih sejumlah komponen utama yang paling penting, sementara sebagian besar variasi dalam data tetap dipertahankan. Dengan demikian, PCA dapat membantu mengatasi masalah dimensi tinggi, mempercepat waktu komputasi, dan mengurangi overfitting, sehingga meningkatkan performa model secara keseluruhan.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)  # Pilih jumlah komponen utama yang diinginkan
```

```python
df_pca = pca.fit_transform(dataframe_scaled)
```

```python
df_pca = pd.DataFrame(
    df_pca, columns=["Principal Component 1", "Principal Component 2"]
)
```
