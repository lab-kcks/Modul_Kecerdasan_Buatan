# Unsupervised Learning

#### Apa itu Unsupervised Learning? 
*Unsupervised Machine Learning* adalah jenis *machine learning* yang melibatkan penemuan pola dalam data yang tidak berlabel. Tidak seperti *Supervised Learning*, di mana model dilatih pada data yang dipasangkan dengan jawaban yang benar, algoritma *Unsupervised Learning* harus bekerja dengan kumpulan data yang tidak memiliki label - artinya, tidak ada instruksi eksplisit tentang pola apa yang harus ditemukan. Algoritma ini berusaha menemukan struktur yang ada dalam data, seperti mengelompokkan titik-titik data yang serupa dalam *clustering* atau menentukan distribusi data dalam *dimensionality reduction*.

#### Task apa saja yang dapat dilakukan dengan Unsupervised Learning?
Dalam penerapannya, *Unsupervised Learning* dapat menyelesaikan banyak jenis *task* diantaranya:
- Clustering
- Dimensionality reduction
- Anomaly Detection
- Word Embeddings
- Dll

Pada modul ini, kita akan berfokus pada salah satu *task* yaitu *Clustering*.

**Tujuan**  
Untuk mengidentifikasi kelompok pelanggan yang berbeda dengan perilaku dan preferensi pembelian yang serupa untuk menciptakan strategi pemasaran yang ditargetkan.

**Solusi**  
Perusahaan e-commerce ini dapat menggunakan algoritma *Clustering* untuk mengelompokkan basis pelanggannya ke dalam kelompok-kelompok yang berbeda. Prosesnya dimulai dengan pengumpulan dan *preprocessing* data pelanggan untuk memastikan data tersebut bersih dan diformat dengan benar. Perusahaan kemudian menggunakan algoritma clustering, seperti K-Means, yang cocok untuk mengelompokkan kumpulan data yang besar. Algoritme ini mengelompokkan pelanggan ke dalam beberapa klaster berdasarkan kesamaan pola dan perilaku pembelian mereka. Setiap klaster mewakili segmen dengan karakteristik yang sama, sehingga memungkinkan perusahaan untuk menyesuaikan strategi pemasarannya secara efektif. Kualitas cluster ini kemudian dievaluasi menggunakan metrik internal seperti Silhouette Coefficient untuk memastikan bahwa segmentasi tersebut bermakna dan dapat ditindaklanjuti.

## 1. Clustering

#### Apa itu Clustering?  

Clustering adalah salah satu jenis metode *Unsupervised Learning* yang bertujuan untuk membagi populasi atau titik data ke dalam beberapa kelompok sehingga titik-titik data lebih mirip dengan titik-titik data lain dalam kelompok yang sama dan berbeda dengan titik-titik data dalam kelompok lain. Pada dasarnya, metode ini merupakan pengumpulan objek berdasarkan kemiripan dan ketidakmiripan di antara mereka.

Sebagai contoh, perhatikan titik-titik data pada grafik di bawah ini. Dari grafik tersebut, kita dapat membedakan kelompok-kelompok dari tiap titik data, dan kita dapat mengidentifikasi bahwa ada 3 cluster pada gambar di bawah ini.

#### Jenis Metode Clustering
**Centroid-based**
- Metode ini memisahkan tiap titik data berdasarkan beberapa centroid pada data. Tiap titik data dikelompokkan ke dalam sebuah cluster berdasarkan jarak kuadrat mereka dari centroid. Algoritma centroid-based merupakan algoritma yang paling sering digunakan. Contohnya adalah algoritma K-Means.

**Hierarchical-based (Connectivity-based)**
- Metode ini didasarkan pada gagasan bahwa suatu objek cenderung lebih berhubungan dengan objek di sekitarnya daripada dengan objek yang lebih jauh. Metode ini membangun sebuah pohon dari cluster sehingga semuanya terorganisir dari atas ke bawah. Pada awalnya, setiap titik data dianggap sebagai cluster individual. Pada setiap iterasi, cluster yang mirip akan digabungkan dengan cluster lainnya hingga membentuk satu cluster besar atau K cluster.

**Density-based**
- Metode ini mencari wilayah-wilayah di mana titik data berkumpul dengan rapat, dan daerah-daerah tersebut dianggap sebagai cluster. Algorima ini mampu untuk mengenali cluster dengan bentuk yang tidak teratur atau tidak biasa, dan juga mengabaikan titik-titik data yang jauh dari cluster utama atau outlier. 
Data is grouped by areas of high concentrations of data points surrounded by areas of low concentrations of data points. Basically the algorithm finds the places that are dense with data points and calls those clusters.

#### Algoritma dalam Clustering

Dalam penerapan dari metode-metode tersebut, terdapat beberapa algoritma yang dapat digunakan dalam melakukan *Clustering*. Berikut merupakan beberapa algoritma yang sering digunakan:
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
- K-Means Clustering
- Mini-Batch K-Means Clustering
- Mean Shift Clustering
- OPTICS (Ordering Points To Identify the Clustering Structure)

- dll

### 1.1 DBSCAN

#### Apa itu DBSCAN?
DBSCAN (*Density-Based Spatial Clustering of Applications with Noise*) merupakan algoritma *clustering* [non-parametrik](https://deepai.org/machine-learning-glossary-and-terms/non-parametric-model#:~:text=What%20are%20Non%2Dparametric%20Models,fixed%20as%20a%20discrete%20number.) yang berbasis pada kepadatan. Hal ini berarti algoritma ini tidak mengasumsikan distribusi statistik tertentu pada input data. DBSCAN bekerja dengan mengidentifikasi *core sample* di wilayah dengan kepadatan tinggi dan memperluas cluster dari *core sample* ini. Algoritma ini sangat cocok untuk data yang berisi cluster dengan kepadatan yang sama.

#### Istilah dalam DBSCAN
Sebelum memasuki langkah-langkah kerja DBSCAN, ada beberapa istilah yang perlu kita pahami:
- ***Epsilon***: Dalam konteks algoritma DBSCAN, epsilon adalah ukuran jarak untuk menentukan *neighborhood* atau *neighbor* dari sebuah sebuah titik. Epsilon merupakan radius dari lingkaran yang digambar mengelilingi titik yang sedang dipertimbangkan. Jadi, epsilon adalah nilai yang merepresentasikan jarak di sekitar titik tertentu yang dianggap sebagai *neighborhood*nya.

- ***Neighborhood***: *Neighborhood* adalah area yang mencakup semua titik-titik data yang menjadi *neighbor* dari sebuah titik. *Neighborhood* ditentukan oleh sebuah lingkaran dengan radius *epsilon*, di mana titik pusatnya adalah titik data yang sedang kita evaluasi.

- ***Neighbor***: *Neighbor* merupakan titik-titik data yang terletak dalam *neighborhood* dari titik pusat yang sedang dievaluasi.

- ***Min_sample***: min_sample, atau dalam beberapa sumber disebut sebagai minPts, adalah batas minimal jumlah titik yang diinginkan dalam *neighborhood* sebuah titik. Nilai ini merupakan ambang batas terendah untuk jumlah titik yang harus ada di *neighborhood* titik tersebut agar bisa dianggap sebagai *core sample* dalam proses pembentukan kluster. Misalnya, jika kita mengambil z = 4, hal ini berarti bahwa setidaknya harus ada empat titik/*neighbor* dalam *neighborhood* tersebut untuk membentuk sebuah kluster.

#### Cara Kerja DBSCAN

1. **Menentukan *Core Point*** 
    - Berdasarkan *epsilon* dan *min_samples*, identifikasi semua *core point* dalam dataset, yaitu titik-titik yang memiliki setidaknya *min_samples* titik lain dalam *neighborhood* mereka.
<div align="center">
    <img src="Images/DBSCAN Work 1.png" height="400" alt="DBSCAN Work 1">
</div>

2. **Membentuk *Cluster* dari *Core Point*** 
    - Pilih sebuah *core point* yang belum dikunjungi dan buat *cluster* di sekelilingnya. Secara rekursif tambahkan semua *neighbor* yang berada dalam *neighborhood* dan langsung dapat dijangkau secara densitas dari titik inti ini ke dalam *cluster*.

3. **Menggabungkan *Cluster* Tetangga** 
    - Jika salah satu *neighboor* juga merupakan *core point*, gabungkan *neighborhood* mereka dengan *cluster* saat ini.
<div align="center">
    <img src="Images/DBSCAN Work 2.png" height="400" alt="DBSCAN Work 2">
</div>

4. **Masukkan Titik Perbatasan ke Kluster**
    - Masukkan *non core point* yang berada dalam *neighborhood* dari *core point* ke *cluster* dari *core point* tersebut.
<div align="center">
    <img src="Images/DBSCAN Work 3.png" height="400" alt="DBSCAN Work 3">
</div>

5. **Ulangi untuk semua *Core Point* sehingga terbentuk *cluster-cluster* yang terpisah.**

6. **Identifikasi *Noise*** 
    - Setelah semua titik telah dikunjungi, labeli titik-titik yang tidak menjadi bagian dari *cluster* manapun sebagai *noise* atau data yang tidak termasuk dalam *cluster*. Titik ini juga dapat diartikan sebagai *outliers*.

#### Kelebihan DBSCAN

1. Tidak Perlu Menentukan Jumlah Cluster: DBSCAN tidak mengharuskan seseorang untuk menentukan jumlah cluster sebelumnya, yang lebih menguntungkan dibandingkan algoritma seperti K-Means di mana jumlah ini perlu ditentukan terlebih dahulu.

2. Dapat Menangani Bentuk *Cluster* yang Sembarang: Algoritma ini bekerja dengan baik dengan data yang berisi *cluster* dengan bentuk yang berubah-ubah, berbeda dengan banyak algoritma yang mengasumsikan *cluster* berbentuk bola.

3. Menangani *Noise*: DBSCAN memiliki pengertian tentang noise dan kuat terhadap outlier, yang berarti DBSCAN dapat mengidentifikasi dan mengecualikan outlier selama proses pengelompokan.

#### Kekurangan DBSCAN

1. Permsalahan Kepadatan yang Bervariasi: DBSCAN mungkin tidak bekerja dengan baik pada dataset dengan perbedaan kepadatan yang besar. Hal ini karena mungkin sulit untuk memilih kombinasi *epsilon-min_samples* yang sesuai dengan semua *cluster*.

2. Kesulitan dalam Memilih Epsilon: Memilih nilai yang cocok untuk *epsilon* dapat menjadi sulit, terutama jika struktur data tidak dipahami dengan baik.

3. Non-deterministik untuk Titik Perbatasan: Algoritma ini tidak sepenuhnya [deterministik](https://www.sv-europe.com/blog/what-is-the-difference-between-the-various-types-of-statistical-models/#:~:text=A%20deterministic%20model%20is%20a,always%20generate%20the%20same%20output.), terutama untuk titik-titik perbatasan yang dapat dicapai dari lebih dari satu klaster. Titik-titik ini dapat berada di salah satu cluster yang berdekatan.

#### Penerapan Coding

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)

X = StandardScaler().fit_transform(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
import numpy as np

from sklearn import metrics
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3, min_samples=10).fit(X)
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```

### 1.2 K-Means

*K-Means Clustering* adalah algoritma *Unsupervised Learning* yang digunakan secara luas untuk mengelompokkan sekumpulan data tanpa label ke dalam klaster-klaster yang berbeda. Tujuan utama dari K-Means adalah untuk membagi $n$ data observasi menjadi $k$ cluster, di mana setiap data menjadi anggota cluster dengan rata-rata terdekat, yang berfungsi sebagai pusat *cluster* atau *centroid*.

#### Istilah dalam K-Means
Sebelum memasuki langkah-langkah kerja DBSCAN, ada beberapa istilah yang perlu kita pahami:

- ***centroid***:  Titik pusat dari setiap klaster. Titik ini didapatkan dengan menghitung rata-rata (*mean*) dari tiap dimensi data.

- **$k$**: Parameter yang ditentukan pengguna yang menentukan jumlah cluster yang harus ditemukan oleh algoritma.

- ***Euclidean Distance***: Jarak garis lurus antara dua titik dalam ruang Euclidean. Dapat dihitung dengan menggunakan rumus:  

    $d(P, Q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + \cdots + (p_n - q_n)^2}$

- ***Manhattan Distance***: Jumlah perbedaan absolut antar dua titik pada koordinat kartesian. Dapat dihitung dengan menggunakan rumus:

    $D(P, Q) = |p_1 - q_1| + |p_2 - q_2| + \cdots + |p_n - q_n|$

#### Cara Kerja K-Means
1. Memilih $k$ *centroid* dari data. Dapat didapatkan dengan pemilihan acak, ataupun dengan metode lainnya seperti ***Elbow Method***.

2. Memasukkan setiap titik data ke *cluster* dari *centroid* dengan jarak terdekat, yang dapat dihitung dengan menggunakan rumus *Euclidean Distance* ataupun *Manhattan Distance*.

3. Hitung ulang posisi *centroid* dengan menghitung rata-rata dari tiap titik data pada suatu *cluster*.

4. Ulangi langkah-langkah sebelumnya hingga tidak ada perubahan yang signifikan dari posisi *centroid*.


#### Kelebihan K-Means
1. Kemudahan Interpretasi: Sangat mudah untuk memahami output dari K-Means, sehingga mudah untuk dijelaskan kepada para pemangku kepentingan.

2. Skalabilitas: K-Means dapat menangani kumpulan data yang besar secara efisien.

3. Konvergensi Terjamin: Algoritmanya dijamin akan konvergen dan memberikan hasil.

4. Kesederhanaan: Algoritme ini mudah diimplementasikan dan diterapkan pada berbagai macam data.

5. Fleksibilitas: Dapat digunakan dengan berbagai jenis data dan dalam domain yang berbeda.

#### Kekurangan K-Means
1. Perlu Menentukan Jumlah Cluster Sebelumnya: Memilih jumlah cluster yang tepat secara tepat dapat menjadi sulit tanpa pemahaman domain (pemahaman data).

2. Sensitif Terhadap *Outliers*: *Outliers* dapat menggeser posisi *centroid*, sehingga menyebabkan pengelompokan yang kurang akurat.

3. *Local Minima*: Algoritma ini dapat konvergen ke minimum lokal, yang mungkin bukan merupakan solusi optimal.

4. Ketergantungan pada Inisialisasi: *Cluster* akhir sangat bergantung pada posisi awal centroid ditempatkan.

5. Keterbatasan dengan Bentuk Cluster: K-Means mengasumsikan cluster berbentuk bola dan berukuran sama, yang mungkin tidak selalu terjadi pada data dunia nyata.

#### Elbow Method
Elbow method merupakan salah satu metode yang digunakan untuk menentukan jumlah cluster yang optimal dalam sebuah dataset. Metode ini melibatkan plotting variasi yang dijelaskan sebagai fungsi dari jumlah klaster dan memilih siku pada kurva sebagai jumlah klaster yang akan digunakan.

Cara kerjanya adalah dengan mencari WCSS (Within-Cluster Sum of Square) yaitu jumlah dari kuadrat jarak antara titik-titik dalam sebuah cluster dan centroid cluster tersebut. Lalu, memplot WCSS sebagai fungsi dari jumlah cluster. Grafik ini sering kali menunjukkan pola berbentuk siku (elbow) di mana WCSS akan menurun seiring bertambahnya jumlah cluster.

Jumlah cluster yang dipilih adalah saat dimana nilai WCSS setelah titik siku tidak menunjukkan perubahan yang signifikan.

#### Penerapan Coding

```python
# Import Libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate synthetic data for demonstration purposes
# X -> data
# y -> label
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
print(X[:10])
print(y[:10])
```

```python
plt.figure(figsize=(12, 5))
plt.scatter(X[:, 0], X[:, 1], s=20)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer

print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(X)
Elbow_M.show()
```

Dari hasil plottingan elbow method di atas, dapat disimpulkan bahwa jumlah cluster optimal untuk data ini adalah 4.

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=4,  n_init='auto')
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
# Plotting the clusters and their centroids
plt.figure(figsize=(12, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### 1.3 Mini-Batch K-Means

Mini-Batch K-Means adalah varian dari algoritma K-Means tradisional yang dirancang khusus untuk bekerja dengan dataset besar. Algoritme K-Means tradisional memproses seluruh dataset dalam setiap iterasi, yang dapat menjadi 'mahal' secara komputasi. Untuk mengatasi hal ini, Mini-Batch K-Means menggunakan sebagian kecil dari seluruh titik data yang acak dan berukuran tetap untuk melakukan pembaruan *centroid* di setiap iterasi.

#### Cara Kerja Mini-Batch K-Means
1. Memilih $k$ *centroid* dari data. Dapat didapatkan dengan pemilihan acak, ataupun dengan metode lainnya.

2. Memasukkan setiap titik data ke *cluster* dari *centroid* dengan jarak terdekat, yang dapat dihitung dengan menggunakan rumus *Euclidean Distance* ataupun *Manhattan Distance*.

3. Berbeda dengan K-means tradisional yang menghitung ulang posisi *centroid* berdasarkan rata-rata seluruh dataset, Mini-Batch K-Means menghitung ulang posisi *centroid* berdasarkan rata-rata dari dalam batch kecil yang acak.

4. Ulangi langkah-langkah sebelumnya hingga tidak ada perubahan yang signifikan dari posisi *centroid*.

#### Kelebihan Mini-Batch K-Means
- Dikarenakan hanya menghitung sebagian kecil dataset pada tiap iterasi, Mini-Batch K-Means lebih cepat daripada algoritme K-Means tradisional, sehingga lebih cocok untuk kumpulan data yang besar.

#### Kekurangan Mini-Batch K-Means
- Mini-Batch K-Means mungkin tidak seakurat/se-konvergen K-Means terdisional dikarenakan tidak menggunakan seluruh dataset.


#### Penerapan Coding

```python
# Import Libraries
from sklearn.cluster import MiniBatchKMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate synthetic data for demonstration purposes
# X -> data
# y -> label
X, y = make_blobs(n_samples=1000, centers=4, cluster_std=0.60, random_state=0)
```

```python
print(X[:10])
print(y[:10])
```

```python
plt.figure(figsize=(12, 5))
plt.scatter(X[:, 0], X[:, 1], s=20)
```

```python
# Apply Mini-Batch K-Means clustering
kmeans = MiniBatchKMeans(n_clusters=4,  n_init='auto')
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
# Plotting the clusters and their centroids
plt.figure(figsize=(12, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('Mini-Batch K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### 1.4 K-Means VS Mini-Batch K-Means

```python
import numpy as np

from sklearn.datasets import make_blobs

np.random.seed(0)

batch_size = 45
centers = [[1, 1], [-1, -1], [1, -1]]
n_clusters = len(centers)
X, labels_true = make_blobs(n_samples=1000000, centers=centers, cluster_std=0.7)
```

```python
import time

from sklearn.cluster import KMeans

k_means = KMeans(init="k-means++", n_clusters=3, n_init=10)
t0 = time.time()
k_means.fit(X)
t_batch = time.time() - t0
```

```python
from sklearn.cluster import MiniBatchKMeans

mbk = MiniBatchKMeans(
    init="k-means++",
    n_clusters=3,
    batch_size=batch_size,
    n_init=10,
    max_no_improvement=10,
    verbose=0,
)
t0 = time.time()
mbk.fit(X)
t_mini_batch = time.time() - t0
```

```python
from sklearn.metrics.pairwise import pairwise_distances_argmin

k_means_cluster_centers = k_means.cluster_centers_
order = pairwise_distances_argmin(k_means.cluster_centers_, mbk.cluster_centers_)
mbk_means_cluster_centers = mbk.cluster_centers_[order]

k_means_labels = pairwise_distances_argmin(X, k_means_cluster_centers)
mbk_means_labels = pairwise_distances_argmin(X, mbk_means_cluster_centers)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 3))
fig.subplots_adjust(left=0.02, right=0.98, bottom=0.05, top=0.9)
colors = ["#4EACC5", "#FF9C34", "#4E9A06"]

# KMeans
ax = fig.add_subplot(1, 3, 1)
for k, col in zip(range(n_clusters), colors):
    my_members = k_means_labels == k
    cluster_center = k_means_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], "w", markerfacecolor=col, marker=".")
    ax.plot(
        cluster_center[0],
        cluster_center[1],
        "o",
        markerfacecolor=col,
        markeredgecolor="k",
        markersize=6,
    )
ax.set_title("KMeans")
ax.set_xticks(())
ax.set_yticks(())
plt.text(-3.5, 1.8, "train time: %.2fs\ninertia: %f" % (t_batch, k_means.inertia_))

# MiniBatchKMeans
ax = fig.add_subplot(1, 3, 2)
for k, col in zip(range(n_clusters), colors):
    my_members = mbk_means_labels == k
    cluster_center = mbk_means_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], "w", markerfacecolor=col, marker=".")
    ax.plot(
        cluster_center[0],
        cluster_center[1],
        "o",
        markerfacecolor=col,
        markeredgecolor="k",
        markersize=6,
    )
ax.set_title("MiniBatchKMeans")
ax.set_xticks(())
ax.set_yticks(())
plt.text(-3.5, 1.8, "train time: %.2fs\ninertia: %f" % (t_mini_batch, mbk.inertia_))

# Initialize the different array to all False
different = mbk_means_labels == 4
ax = fig.add_subplot(1, 3, 3)

for k in range(n_clusters):
    different += (k_means_labels == k) != (mbk_means_labels == k)

identical = np.logical_not(different)
ax.plot(X[identical, 0], X[identical, 1], "w", markerfacecolor="#bbbbbb", marker=".")
ax.plot(X[different, 0], X[different, 1], "w", markerfacecolor="m", marker=".")
ax.set_title("Difference")
ax.set_xticks(())
ax.set_yticks(())

plt.show()
```

## 2. Evaluation Metrics for Clustering

Dalam mengevaluasi sebuah model *clustering*, terdapat beberapa pendekaatan yang dapat dilakukan. Pendekatan-pendekatan tersebut dapat dibagi menjadi dua kelompok utama yaitu *Internal Evaluation* dan *External Evaluation*. Kedua hal ini yang akan kita bahas pada bagian berikut ini.

### 2.1. Internal Evaluation
Evaluasi Internal dalam model *clustering* melibatkan penilaian kualitas pengelompokan dengan memeriksa data yang telah dikelompokkan itu sendiri. Hal ini biasanya mencakup penghitungan jarak-*cluster* antar dan intra-*cluster*, yang bertujuan untuk mendapatkan kesamaan yang tinggi di dalam *cluster* dan kesamaan yang rendah di antara *cluster* yang berbeda.

Terdapat beberapa teknik yang digunakan untuk melakukan *Internal Evaluation*. Beberapa diantaranya akan dibahas pada bagaian berikut ini.

#### a. Sillhouette Coefficient
Koefisien siluet didapatkan dengan menghitung rata-rata jarak intra-cluster dan rata-rata jarak antar-cluster dari setiap titik data.

>*Rumus*:  
>
>$\text{silhouette coefficient} = \frac{b - a}{\max(a, b)}$
>
>*keterangan*:
>- $a$ = jarak rata-rata antara titik data saat ini dan semua titik data lainnya dalam *cluster* yang sama.
>- $b$ = jarak rata-rata antara titik data saat ini dan semua titik data lainnya dalam *cluster* terdekat berikutnya.

Koefisien siluet bervariasi antara -1 hingga 1, dengan -1 menunjukkan bahwa titik data tidak dimasukkan ke dalam klaster yang tepat, 0 menunjukkan bahwa klaster-klaster tersebut saling tumpang tindih, dan 1 menunjukkan bahwa klaster tersebut padat dan terpisah dengan baik.

*code*:

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn import datasets
X, y = datasets.load_iris(return_X_y=True)
```

```python
import numpy as np
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=3, random_state=1, n_init='auto').fit(X)
labels = kmeans_model.labels_
metrics.silhouette_score(X, labels, metric='euclidean')
```

#### b. Davies-Bouldin Index
*Davies-Bouldin Index* dapat dihitung dengan rumurs berikut:

>*rumus*:
>
>$\text{Davies-Bouldin Index} = \frac{1}{c} \sum_{i=1}^{c} \max_{j \neq i} \left( \frac{\sigma_i + \sigma_j}{d(c_i, c_j)} \right)$
>
>*keterangan*:
>
>- $c$ = Jumlah *cluster*
>- $c_i$ = *centroid* dari *cluster* $i$
>- $d(c_i, c_j)$ = jarak antara dua *cluster*
>- $\sigma_i$ = rata-rata jarak dari semua titik data pada *cluster* $i$ ke $c_i$

Model yang memberikan jarak intra-klaster yang rendah dan jarak antar-klaster yang tinggi (metrik yang diinginkan) menghasilkan Indeks Davies-Bouldin yang rendah. Maka dari itu, semakin rendah Indeks Davis-Bouldin, semakin baik modelnya.

*code*:

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
kmeans = KMeans(n_clusters=3, random_state=1, n_init='auto').fit(X)
labels = kmeans.labels_
davies_bouldin_score(X, labels)
```

#### c. Calinski-Harabasz Index
Indeks Calinski-Harabasz (juga disebut kriteria rasio varians) adalah rasio antara sebaran antar-*cluster* dan sebaran intra-*cluster* untuk semua *cluster*. Index ini dapat dihitung dengan rumus sebagai berikut:

>*rumus*:  
>
>$CHI = \frac{trace(B_c)}{trace(W_c)} * \frac{n_E - c}{c - 1}$
>
>*keterangan*:
>- $c$ = jumlah *cluster*
>- $n_E$ = ukuran dataset $E$
>- $trace(B_c)$ = trace dari matriks sebaran antar-*cluster*
>- $trace(W_c)$ = trace dari matriks sebaran inter-*cluster*

Oleh karena itu, CHI diestimasikan berdasarkan jarak titik data dalam sebuah *cluster* dari pusat *cluster* dan jarak semua pusat *cluster* dari pusat global. Semakin tinggi nilai CHI, semakin baik model pengelompokannya.

*code*:

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn import datasets
X, y = datasets.load_iris(return_X_y=True)
```

```python
import numpy as np
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=3, random_state=1, n_init='auto').fit(X)
labels = kmeans_model.labels_
metrics.calinski_harabasz_score(X, labels)
```

### 2.2. External Evaluation
Evaluasi Eksternal, didasarkan pada data yang tidak digunakan untuk pengelompokan. Hal ini dapat mencakup membandingkan hasil pengelompokan dengan tolok ukur eksternal atau ground truth. Pada dasarnya, evaluasi ini mengukur seberapa baik algoritma *cluster* sesuai dengan label yang sudah ada atau klasifikasi yang sudah diketahui.

Terdapat beberapa teknik yang digunakan untuk melakukan *Internal Evaluation*. Beberapa diantaranya akan dibahas pada bagaian berikut ini.

#### a. Rand Index
Indeks Rand (RI) mengukur persentase prediksi yang benar di antara semua prediksi. Untuk menghitungnya, kita dapat menggunakan rumus sebagai berikut:

>*rumus*:  
>
>$RI = \frac{TP + TN}{TP + TN + FP + FN}$
>
>*keterangan*:
>
>- $TP$ = jumlah *True Positive*
>- $TN$ = jumlah *True Negative*
>- $FP$ = jumlah *False Positive*
>- $FN$ = jumlah *False Negative*

Indeks Rand berkisar antara 0 hingga 1, dengan 1 menunjukkan model pengelompokan yang baik.

*code*:

```python
from sklearn import metrics
labels_true = [0, 0, 0, 1, 1, 1]
labels_pred = [0, 0, 1, 1, 2, 2]
metrics.rand_score(labels_true, labels_pred)
```

#### b. Adjusted Rand Index

*Adjusted Rand Index*(ARI) merupakan RI yang telah diperbaiki untuk peluang. ARI menyesuaikan RI dengan mempertimbangkan kemiripan yang diharapkan dari semua perbandingan berpasangan antar-*cluster* dan dengan demikian memberikan skor yang lebih kuat terhadap jumlah pengelompokan dan ukuran dataset. Untuk mendapatkan ARI, kita dapat menggunakan rumus berikut:

>*rumus*:
>
>$ARI = \frac{ \text{Index} - \text{Expected Index} }{ \text{Max Index} - \text{Expected Index} }$
>
>*keterangan*:
>
>- $\text{Index}$ = Rand Index (RI) = $\frac{TP + TN}{TP + FP + FN + TN}$
>- $\text{Expected Index}$ = RI yang diharapkan pada peluang acak
>- $\text{Max Index}$ = Nilai maksimum yang mungkin dari RI

ARI memiliki rentang nilai yang mencakup nilai negatif. Nilai yang mendekati 1 menunjukkan kemiripan yang tinggi antar dua *clustering* (*ground-truth* - *prediction*), nilai yang mendekati 0 menunjukkan pelabelan acak, dan nilai negatif menunjukkan pelabelan independen yang kurang mirip dari yang diharapkan secara *random*.

*code*:

```python
from sklearn import metrics
labels_true = [0, 0, 0, 1, 1, 1]
labels_pred = [0, 0, 1, 1, 2, 2]
metrics.adjusted_rand_score(labels_true, labels_pred)
```

#### c. Normalized Mutual Information (NMI)
*Normalized Mutual Information* adalah normalisasi skor dari Mutual Information (MI) untuk menskalakan hasil antara 0 (tidak ada informasi mutual) dan 1 (korelasi sempurna). Hal ini digambarkan pada rumus berikut:

>*rumus*:
>
>$NMI(\Omega, C) = \frac{I(\Omega; C)}{\frac{H(\Omega) + H(C)}{2}}$
>
>*keterangan*:
>
>- $I(\Omega; C)$ = *Mutual Information* antara *clustering* $\Omega$ dan $C$
>- $H(\Omega)$ dan $H(C)$ = Nilai entropi dari *clustering* $\Omega$ dan $C$

*code*:

```python
from sklearn import metrics
labels_true = [0, 0, 0, 1, 1, 1]
labels_pred = [0, 0, 1, 1, 2, 2]

metrics.normalized_mutual_info_score(labels_true, labels_pred)  
```
